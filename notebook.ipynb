# Deepfake Audio Detection Pipeline

This notebook demonstrates the full reproducible pipeline from feature extraction to training and inference.

```
import os
import glob
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score
from scripts.extract_features import extract_audio_features
import joblib

print("Libraries imported successfully.")
```

## 1. Feature Extraction (Sample)

```
sample_audio = glob.glob("data/for-norm/for-norm/*/*/*.wav")[0]
features = extract_audio_features(sample_audio)
print(f"Extracted 90-dimensional feature vector for {os.path.basename(sample_audio)}\nShape: {np.array(features).shape}")
```

## 2. Load Model and Predict

```
model = joblib.load("models/lgbm_model.pkl")
pred = model.predict(np.array(features).reshape(1, -1))[0]
prob = model.predict_proba(np.array(features).reshape(1, -1))[0]
print(f"Prediction: {"DEEPFAKE" if pred == 1 else "GENUINE"} (Confidence: {prob[pred]:.4f})")
```
